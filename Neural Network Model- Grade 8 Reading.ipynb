```python
# Importing libraries.

import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
avg_scores = pd.read_csv('input_data/naep.csv')
poverty_income = pd.read_csv('input_data/poverty_income.csv')
unemploy_crime = pd.read_csv('input_data/unemployment_crime.csv')
```

```python
unemploy_crime.info()
```

```python
merged_df = pd.merge(avg_scores, poverty_income, how = 'left' , left_on =['state', 'year'], right_on=['state', 'year'])
merged_df = pd.merge(merged_df, unemploy_crime, how = 'left' , left_on =['state', 'year'], right_on=['state', 'year'])
merged_df
```

## Neural Network Testing - Grade 8 Reading

```python
reading_merged_df = merged_df.loc[(merged_df['subject'] == 'Reading')]
read_8_merged_df = reading_merged_df.loc[(merged_df['grade'] == 8)]
read_8_merged_df
```

```python
read_8_merged_df = read_8_merged_df.drop(columns = 'year')
read_8_merged_df = read_8_merged_df.drop(columns = 'state')
read_8_merged_df = read_8_merged_df.drop(columns = 'subject')
read_8_merged_df = read_8_merged_df.drop(columns = 'grade')
read_8_merged_df = read_8_merged_df.drop(columns = 'violent_crime_rate')
read_8_merged_df = read_8_merged_df.drop(columns = 'property_crime_rate')
read_8_merged_df
```

```python
read_8_merged_df.isnull().sum()
```

```python
read_8_merged_df.describe()
```

```python
bins = [0, 250.9, 260.9, 270.9, 280]
group_names = [0, 1, 2, 3]
read_8_merged_df['new_col'] = pd.cut(read_8_merged_df['avg_score'],bins, labels=group_names, include_lowest=True)
```

```python
read_8_merged_df = read_8_merged_df.drop(columns = 'avg_score')
read_8_merged_df
```

```python
x = read_8_merged_df.drop(columns = 'new_col').values
y = read_8_merged_df['new_col'].values
```

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split

y = tf.keras.utils.to_categorical(y)
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
x_scaler = scaler.fit(x_train)

# Scale the data
x_train_scaled = x_scaler.transform(x_train)
x_test_scaled = x_scaler.transform(x_test)
```

```python
input_features = len(x_train_scaled[0])
hidden_layer1 = 50
hidden_layer2 = 35
# hidden_layer3 = 10

nn = tf.keras.models.Sequential()

nn.add(tf.keras.layers.Dense(units = hidden_layer1, input_dim = input_features, activation = 'relu'))

nn.add(tf.keras.layers.Dense(units = hidden_layer2, activation = 'relu'))

# nn.add(tf.keras.layers.Dense(units = hidden_layer3, activation = 'relu'))

nn.add(tf.keras.layers.Dense(units = 4, activation = 'softmax'))

nn.summary()
```

```python
nn.compile(loss="categorical_crossentropy", optimizer= "adam", metrics=['accuracy'])
```

```python
fit_model = nn.fit(x_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(x_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
